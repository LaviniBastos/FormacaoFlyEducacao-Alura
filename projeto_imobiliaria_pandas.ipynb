```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'

dados = pd.read_csv(url)
dados
```

```python
dados.isnull().value_counts()
```

```python
dados = dados.fillna(0)
```

```python
dados
```

```python
# Removendo essa coluna

dados.drop(7, axis=0, inplace=True)
```

```python
dados
```

```python
# Filtro buscando alunos aprovados
# precisamos trocar o true or false para 0 = reprovado e 1 aprovado

dados['Aprovado'].value_counts()
```

```python

dados['Aprovado'] = dados['Aprovado'].astype(int)
```

```python
aprovados = dados['Aprovado'] == 1
dados[aprovados]
```

```python
# Salvando os aprovados em um arquivo
alunos_aprovados = dados[aprovados]

alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
```

```python
alunos_aprovados
```

```python
# Atividades extras
# 1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras.
# Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso,
# crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

dados['Nota_extra'] = dados['Notas'] * 0.40
dados
```

```python
# 2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.

dados['Notas_Finais'] = dados['Notas'] + dados['Nota_extra']
dados
```

3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:

True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);

False: caso o aluno esteja reprovado (nota final deve ser menor que 6).

```python
dados['Aprovado_final'] = dados['Notas_Finais'].apply(lambda x:'True' if x > 6 else 'False')
dados
```

4) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

```python
#Validando as mudanças que ocorreram

dados['Aprovado_final'].value_counts()
```

```python
#Validando as mudanças que ocorreram

dados['Aprovado'].value_counts()
```

```python
# Para ter certeza vou converter as duas colunas para um tipo só

dados['Aprovado'] = dados['Aprovado'].astype(str)
```

```python
filtro_1 = dados['Aprovado'] == '0'
filtro_2 = dados['Aprovado_final'] == 'True'

filtro_completo = filtro_1 & filtro_2


resultado = dados[filtro_completo][['Nome', 'Aprovado', 'Aprovado_final']]

resultado
```
